```python
from sklearn.svm import SVC
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

##### Read data into python dataframes

##### Read data into pandas dataframe
read.csv reads train and test data into respective pandas dataframes 

```python
Train_data = pd.read_csv('http://www.amlbook.com/data/zip/features.train', delim_whitespace=True, header=None)
Test_data = pd.read_csv('http://www.amlbook.com/data/zip/features.test', delim_whitespace=True, header=None)
```

##### Taking only those rows which have label either 1 or 5

```python
Train_data = Train_data.loc[(Train_data.iloc[:, 0] == 1) | (Train_data.iloc[:, 0] == 5)]
Test_data = Test_data.loc[(Test_data.iloc[:, 0] == 1) | (Test_data.iloc[:, 0] == 5)]
```

##### Updating labels :  '1' becomes '+1' and '5' becomes '-1' 

```python
Train_data.loc[Train_data.iloc[:, 0] == 1, 0] = +1
Train_data.loc[Train_data.iloc[:, 0] == 5, 0] = -1
Test_data.loc[Test_data.iloc[:, 0] == 1, 0] = +1
Test_data.loc[Test_data.iloc[:, 0] == 5, 0] = -1
```

##### Dividing the train and test data into features and labels using iloc as first column of train and test dataframes
Using .values to convert dataframes into Numpy ndarray

```python
X_train = Train_data.iloc[:,1:].values
X_test = Test_data.iloc[:,1:].values
y_train = Train_data.iloc[:,0].values
y_test = Test_data.iloc[:,0].values
```

##### Data preprocessing : standardization

```python
X_train = StandardScaler().fit_transform(X_train)
X_test = StandardScaler().fit_transform(X_test)
```

##### Part a) Linear kernel with Soft margin SVM

```python
svm_linear = SVC(kernel='linear', C=0.1)
```

```python
svm_linear.fit(X_train, y_train)
```

```python
print("The test accuracy for svm linear kernel is:",svm_linear.score(X_test, y_test))
```

```python
print("Number of Support vectors for the svm linear kernel is: ",svm_linear.support_vectors_.shape[0])
```
